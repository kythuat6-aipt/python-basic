## 9. Sort()

```python
# inplace mặc định là False
# inplace = True -> lưu lại vào DataFrame | False -> Ko lưu lại

# ascending mặc định là True
# ascending = True -> sắp xếp theo tăng dần | False -> sắp xếp theo giảm dần
df.sort_values('Internet users',  ascending=False, inplace=False)
```

## 11. Merge:
+ dùng để kêt hợp các dataframe lại với nhau dự trên một hoặc nhiều cột
+ left: là bảng bên trái 
+ right: bảng bên phải
+ how: INNER, OUTER, LEFT, RIGHT -> giống JOIN trong SQL
+ on: là trên cột nào

<img src="https://data36.com/wp-content/uploads/2018/08/4-pandas-merge-inner-outer-left-right-1024x771.png" width="400">

```python
df.head()
```

```python
Income_Group = pd.read_excel('./data/income_group.xlsx')
```

```python
Income_Group
```

```python
df_merge = pd.merge(
    left= df,
    right= Income_Group[['Income Group','Income (updated)']],
    how= 'left',
    on= 'Income Group'
)
```

## 13. Group By

```python
df['Income Group'].value_counts()
```

```python
df_new = pd.read_csv('./data/restaurant.csv')
```

```python
df_new.head()
```

```python
# lấy tổng total_bill theo ngày
df_new.groupby('day').sum().reset_index()
```

```python
# lấy ra tổng tiền tip theo ngày và theo giới tính
df_new.groupby(['day', 'sex'])['tip'].sum()
```

```python
# lấy ra tổng và trung bình của tip và total_bill theo ngày
df_new.groupby('day')[['tip', 'total_bill']].agg(['mean', 'sum'])
```

```python
# lây ra trung bình của tip và tổng của total_bill theo ngày
df_new.groupby('day')[['tip','total_bill']].agg({'tip':'mean', 'total_bill':'sum'})
```

## 14. Pivot Table

<img src="https://pandas.pydata.org/docs/_images/reshaping_pivot.png" width="400">

```python
# lấy ra tổng tiền tip theo sex và day
pd.pivot_table(
    data= df_new,
    index= 'sex',
    columns= 'day',
    values= 'tip',
    aggfunc= 'sum'
)
```

```python
df_new
```

```python
pd.pivot_table(
    data= df_new,
    index= 'sex',
    columns= 'time',
    values='tip',
    aggfunc=['min', 'mean', 'max', 'sum']
)
```

## 15. Apply
+ axis = 1: lọc qua từng dòng
+ axis = 0: lọc qua từng cột

```python
def check_tip (row):
    if row['tip'] > 3:
        return 'Hight tip'
    else:
        return 'Low tip'
```

```python
# tạo ra một trường tip type để lưu loại tip
df_new['tip type'] = df_new.apply(check_tip, axis=1)
df_new
```
